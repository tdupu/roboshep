```python
def get_member_by_name(name):
    all_members = client.get_members()['members']
    for member in all_members:
        if member['full_name']==name:
            return member
        
def get_member_by_email(email):
    all_members = client.get_members()['members']
    for member in all_members:
        if member['email']==email:
            return member
        
def get_subscriber_emails(streamname):
    return admin_client.get_subscribers(stream=streamname)['subscribers']

def get_subscriber_ids(streamname):
    emails = get_subscriber_emails(streamname)
    user_ids = []
    for email in emails:
        member = get_member_by_email(email)
        user_id = member['user_id']
        user_ids.append(user_id)
    return user_ids
```

```python
def add_to_admin_streams(user_ids):
    """
    user_ids must be a list of, well, user ids.
    """
    all_streams = admin_client.get_streams()['streams']
    streamname_dicts = [ {'name':stream['name']} for stream in all_streams] 
    result = admin_client.add_subscriptions(
    streams=streamname_dicts,
    principals=user_ids)
    return result
```

```python
shepherd_ids = get_subscriber_ids(SHEPHERD_STREAM_NAME)
print(shepherd_ids)
```

```python
roboshep_id = get_member_by_name('roboshep')['user_id']
print(roboshep_id)
```

```python
ids_to_add = shepherd_ids + [roboshep_id]
add_to_admin_streams(ids_to_add)
```
